```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

xinfadi_pd = pd.read_pickle('xinfadi.pkl')
pro_name = ''巨峰葡萄''
avg = xinfadi_pd[xinfadi_pd['品名'].str.contains(pro_name)]['平均价']
avg = avg.sort_index()
avg = avg.astype(float)

plt.rcParams['font.sans-serif']=['SimHei']
fig  = plt.figure()
fig.set_figwidth(20)
plt.xlabel('发布时间')
plt.ylabel(pro_name)
avg = avg.sort_index()
avg= avg
plt.plot(avg.index, avg)
```
