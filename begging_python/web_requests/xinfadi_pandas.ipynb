```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

xinfadi_pd = pd.read_pickle('xinfadi.pkl')
pro_name = '蜜柚'
avg = xinfadi_pd[(xinfadi_pd['品名'] == pro_name) & (xinfadi_pd['规格'] == '红心')]['平均价']
avg = avg.sort_index()['2022']
avg = avg.astype(float)

plt.rcParams['font.sans-serif']=['SimHei']
fig  = plt.figure()
fig.set_figwidth(20)
plt.xlabel('发布时间')
plt.ylabel(pro_name)
plt.plot(avg.index, avg)
```
