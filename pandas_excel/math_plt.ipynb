```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']  #用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  #用来正常显示负号

x = np.arange(-5.0, 5.0, 0.0001)
y = np.where(np.fabs(x) < (np.pi / 3), np.fabs(np.sin(x)), 0)
plt.title('高数习题1-1 第7题 y = |sin x|')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']  #用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  #用来正常显示负号

x = np.arange(-105.0, 105.0, 0.01)
y = x * np.cos(x)
plt.title('高数习题1-1 第13题(4) y = x cos x')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']  #用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  #用来正常显示负号

x = np.arange(-5.01, 5.01, 0.1)
y = (1 - x )/ (1 + x)
plt.title('高数习题1-1 第13题(5) y = sin2 x')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, y)
plt.show()
```
