```python
import pandas as pd
import xlwings as xw


df_l = pd.read_excel('./xl/tb/店铺引流关键词_每日-2022-09-12.xls', header=7)
df_key = df_l[['关键词', '引导支付金额', '带来的访客数', '引导下单买家数', '引导支付件数']]
df_key = df_key.groupby(['关键词']).sum()
df_key['下单转化率(%)'] = df_key['引导下单买家数'] / df_key['带来的访客数']*100
df_key = df_key.sort_values(by=['带来的访客数', '下单转化率(%)'], ascending=False)
df_key
#xw.view(df_key)
```

```python
df_car = pd.read_csv('./xl/tb/20220912050422_关键词.csv')
df_car = df_car[['词包名/关键词', '点击量', '总成交笔数', '点击转化率']]
df_car.info()
df_car = df_car[df_car['点击量'] > 0]

df_car['下单数'] = df_car['点击量'] * df_car['点击转化率'] / 100

df_car = df_car.groupby(['词包名/关键词']).sum()
df_car['点击转化率'] = df_car['下单数'] / df_car['点击量'] * 100
df_car = df_car.sort_values(by=['点击转化率'], ascending=False)
xw.view(df_car)
```
